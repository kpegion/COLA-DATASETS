```python
import xarray as xr
import numpy as np
import pandas as pd
import glob
```

```python
subx_path='/shared/subx/'
subx_type='hindcast'
subx_var='tas'
subx_lev='2m'
subx_model='EMC-GEFSv12'
subx_dates='*'
```

```python
path=subx_path+subx_type+'/'+subx_var+subx_lev+'/daily/full/'+subx_model+'/'
files_inits=sorted(glob.glob(path+subx_var+'_'+subx_lev+'_'+subx_model+'_'+subx_dates+'e1.daily.nc'))
inits=[pd.to_datetime(i.split('.', 1)[0].split('_')[3],format='%Y%m%d') for i in files_inits]
```

```python
fcstds=xr.open_mfdataset(files_inits,combine='nested',concat_dim='init')
fcstds['init']=inits
fcstds=fcstds.dropna(dim='init')
fcstds
```
